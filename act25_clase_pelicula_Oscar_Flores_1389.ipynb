```python

class Pelicula:
    # El constructor para inicializar los 3 atributos
    def __init__(self, titulo, genero, duracion):
        self.titulo = titulo
        self.genero = genero
        self.duracion = duracion

    # Función para mostrar la información en pantalla
    def mostrardatos(self):
        print(f"--- Información de la Película ---")
        print(f"Título: {self.titulo}")
        print(f"Género: {self.genero}")
        print(f"Duración: {self.duracion} minutos")
        print("-" * 30)

# --- INICIO DEL PROGRAMA INTERACTIVO ---

print("Bienvenido al registro de películas")
print("Por favor, ingresa los datos de dos películas.\n")

# Solicitamos datos para el objeto 1 (pelicula1)
print("Datos de la Película 1:")
nom1 = input("Escribe el título: ")
gen1 = input("Escribe el género: ")
dur1 = input("Escribe la duración (en minutos): ")
# Creamos el primer objeto
pelicula1 = Pelicula(nom1, gen1, dur1)

print("\nDatos de la Película 2:")
# Solicitamos datos para el objeto 2 (pelicula2)
nom2 = input("Escribe el título: ")
gen2 = input("Escribe el género: ")
dur2 = input("Escribe la duración (en minutos): ")
# Creamos el segundo objeto
pelicula2 = Pelicula(nom2, gen2, dur2)

# Usamos la función mostrardatos para ver los resultados
print("\nMostrando resultados finales:\n")
pelicula1.mostrardatos()
pelicula2.mostrardatos()

print("Programa terminado")
print("Creado por el alumno de software pro Oscar Flores NL = 19")
```
